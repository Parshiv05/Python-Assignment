# string operations
* Write a Python program that takes a user first and last name as input and prints them in reverse order with a space between them.

* Use at least two string methods and explain their purpose in the comments.

```python
#Taking first name and last name input from the user
first_name = input("Enter your first name: ").strip() #.strip() It removes any spaces at the beginning and end of any string input
last_name = input("Enter your last name: ").strip()

#Combining and reversing the order of the names
reversed_name = f"{last_name.capitalize()} {first_name.capitalize()}"  #.capitalize() makes the first letter uppercase and the rest lowercase

#Printing the reversed names as a result
print(reversed_name)
```

* Implement a loop that continues to ask the user for a number they enter ''

```python
while True:
    #asking user to enter a number or type 'exit' to quit
    user_input = input("Enter a number (or type 'exit' to quit): ")
    
    #Check if the user wants to exit the loop
    if user_input.lower() == 'exit':
        print("Exiting the program.")
        break
    
    #Attempt to convert the input to a float
    try:
        number = float(user_input)
        
        #Determining whether the number is positive, negative, or zero
        if number > 0:
            print("The number is positive.")
        elif number < 0:
            print("The number is negative.")
        else:
            print("The number is zero.")
    
    #if invalid input
    except ValueError:
        print("Invalid input. Please enter a valid number or type 'exit' to quit.")
```

# Relational and logical operator 

* Create a python script that takes two numbers as input and prints whether both numbers are even odd, or one of each using relational and logical operators

```python
#ask the user to enter the first and second numbers
num1 = int(input("Enter the first number: "))  #Convert the input to an integer
num2 = int(input("Enter the second number: "))  #convert the input to an integer

#Use relational and logical operators to check the parity of the numbers
if num1 % 2 == 0 and num2 % 2 == 0:
    print("Both numbers are even.")
elif num1 % 2 != 0 and num2 % 2 != 0:
    print("Both numbers are odd.")
else:
    print("One number is even and the other is odd.")
```

# For loop and bitwise operators

* write a python program that takes an integer input and prints its binary, ocal, and hexadecimal equivalents using a for loop and bitwise operators

```python
#ask the user to enter an integer
number = int(input("Enter an integer: "))

# Define a function to convert a number to binary, octal, and hexadecimal manually using bitwise operations
def convert_to_bases(num):
    #Create lists to store binary, octal, and hexadecimal representations
    binary = []
    octal = []
    hexadecimal = []

    #Binary conversion using bitwise right shift and mask with 1
    temp = num
    while temp > 0:
        binary.append(str(temp & 1))  #Append the least significant bit
        temp >>= 1  #Right shift by 1 to move to the next bit
    binary.reverse()  #Reverse the list to get the correct binary representation

    #Octal conversion by grouping bits in sets of 3
    temp = num
    while temp > 0:
        octal.append(str(temp & 0b111))  #Mask last three bits
        temp >>= 3  #Right shift by 3 bits
    octal.reverse()  #Reverse to get the correct octal representation

    #Hexadecimal conversion by grouping bits in sets of 4
    hex_map = '0123456789ABCDEF'  #Map to convert values 0-15 to hex digits
    temp = num
    while temp > 0:
        hexadecimal.append(hex_map[temp & 0b1111])  #Mask last four bits
        temp >>= 4  #Right shift by 4 bits
    hexadecimal.reverse()  #Reverse to get the correct hexadecimal representation

    #Print the results
    print(f"Binary: {''.join(binary)}")
    print(f"Octal: {''.join(octal)}")
    print(f"Hexadecimal: {''.join(hexadecimal)}")

#Call the function with the input number
convert_to_bases(number)
```
